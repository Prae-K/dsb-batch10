```python
## 1) ช่วงเวลาที่คนนิยมเดินทาง
#สร้างช่วงเวลาออกเดินทาง
flights_time <- df %>%
  mutate(
    time_of_day = case_when(
      dep_time >= 500 & dep_time < 1200 ~ "Morning",
      dep_time >= 1200 & dep_time < 1700 ~ "Afternoon",
      dep_time >= 1700 & dep_time < 2200 ~ "Evening",
      TRUE ~ "Night"
    )
  )
flights_time %>%
  count(time_of_day)
#ช่วงที่คนนิยมเดินทางมากที่สุดเป็นช่วงเช้า
```

```python
## 2) สายการบินที่ล่าช้า
#กำหนดเกณฑ์ล่าช้า > 5
delay_flight <- df %>%
  filter(arr_delay > 5) %>%
  group_by(carrier) %>%
  summarize(num_flights = n(),
            delay_flights = sum(arr_delay),
            delay_rate = mean(arr_delay)) %>%
  arrange(desc(delay_rate))
delay_flight
##ผู้ให้บริการที่มีค่าเฉลี่ยเวลา delay ที่สุด คือ OO
```

```python
## 3) สายการบินที่ได้รับความนิยมสูงที่สุด
popular_flight <- df %>%
  group_by(carrier) %>%
  summarize(n_flight = n()) %>%
  arrange(desc(n_flight))
head(popular_flight,10)
#สายการบินที่ได้รับความนิยมสูงสุด คือ UA
```

```python
## 4. หาว่า wind gust ส่งผลต่อ flight delay รึป่าว
# join table flights and weather
flights_weather <- df %>%
  left_join(weather, by = c("origin", "year", "month", "day", "hour"))
flights_weather
# groub delay, no delay แล้วหา avg_wind_gust
flights_weather %>%
  select(arr_delay, wind_gust) %>%
  mutate(arr_delay = ifelse(arr_delay > 5, "delay", "no delay")) %>%
  group_by(arr_delay) %>%
  summarize(
    avg_wind_gust = mean(wind_gust, na.rm = TRUE),
    max_wind_gust = max(wind_gust, na.rm = TRUE),
    n_flights = n()
  )
# avg_wind_gust ของ delay > no delay (ทำ hypothesis test ต่อ)
```
