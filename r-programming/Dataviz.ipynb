```r
# ต้นไม้ที่ส่วนสูงมากน้ำหนักก็จะมากด้วย
ggplot(trees, aes(x = Height, y = Volume)) +
  geom_point(size = 2, color = "green") +
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  theme_minimal()
```

```r
avg_chick_weight <- ChickWeight %>%
  group_by(Time, Diet) %>%
  summarise(avg_weight = mean(weight))
ggplot(avg_chick_weight, aes(x = Time, y = avg_weight, color = Diet)) +
  geom_line() +
  theme_minimal()
## น้ำหนักเฉลี่ยของไก่ group 1 น้อยที่สุด group 3 มากที่สุด
```
